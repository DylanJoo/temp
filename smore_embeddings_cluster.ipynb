```python
from sklearn.manifold import TSNE
import numpy as np
from keras.datasets import mnist
from sklearn.datasets import load_iris
from numpy import reshape
import seaborn as sns
import pandas as pd  
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity as cs
import numpy as np
```

```python
def dim_reduct(path, drop=None):
    if drop:
        df = pd.read_csv(f"{path}", index_col=0).drop(drop)
        print(len(df))
        return df.values

    else:
        df = pd.read_csv(f"{path}", index_col=0)
        print(len(df))
        y = df.index == 'cold'
        return df.values, y

def get_index(path, drop=None):
    if drop:
        df = pd.read_csv(f"{path}", index_col=0).drop(drop)
        return df.index

    else:
        df = pd.read_csv(f"{path}", index_col=0)
        return df.index
```

```python
# 20220708
tsne = TSNE(n_components=2, verbose=1, random_state=123)
x, drop = dim_reduct('/content/test/20220708-U-uw.binary.csv')
z = tsne.fit_transform(x) 

sns.scatterplot(x=z[:, 0], y=z[:, 1], s=5)
sns.scatterplot(x=z[0, 0], y=z[0, 1], hue=['pure-cold'], s=100)
```

```python
# 20220708
tsne = TSNE(n_components=2, verbose=1, random_state=123)
x = dim_reduct('/content/test/20220708-U-uw.binary.csv', drop='cold')
kmeans = KMeans(n_clusters=2, random_state=123).fit(x)
z = tsne.fit_transform(x) 
sns.scatterplot(x=z[:, 0], y=z[:, 1], hue=kmeans.labels_, s=5)
```

```python
selected_id = np.argwhere(kmeans.labels_).flatten()
index_name = get_index('/content/test/20220708-U-uw.binary.csv', drop='cold')
index_name[selected_id]
```
